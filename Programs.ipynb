```python
for i in range(5):
    for j in range(i + 1):
        print("#",end= " ")
    print()
```

```python
for i in range(5):
    for j in range(4-i):
        print("#", end= " ")
    print()
```

```python
for i in range(4):
    for j in range(4):
        print("*", end= " ")
    print()
```

```python
iSum = 0
for i in range(1, 101):
    iSum += i
print("The sum of numbers from 1 to 100 is:", iSum)
```

```python
iNum = 7
for i in range(2, iNum):
    if iNum % i == 0:
        print("Not Prime")
        break
    else:
        print("prime")
```
